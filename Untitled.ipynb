```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

#import tensorflow as tf

from sklearn.metrics import confusion_matrix, classification_report
```

```python
data = pd.read_csv('emotions.csv')
```

```python
data.head(20)
```

```python
sample = data.loc[0, 'fft_0_b':'fft_749_b']

plt.figure(figsize=(16, 10))
plt.plot(range(len(sample)), sample)
plt.title("Features fft_0_b through fft_749_b")
plt.show()
```

```python
data['label'].value_counts()
```

```python
label_mapping = {'NEGATIVE': 0, 'NEUTRAL': 1, 'POSITIVE': 2}
```

```python

df = data   
df['label'] = df['label'].replace(label_mapping)   
#y = df['label'].copy()
#X = df.drop('label', axis=1).copy()   
#X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=123)
```

```python
df
```

```python
df.columns
```

```python
X =df[['mean_1_a', 'mean_2_a', 'mean_3_a', 'mean_4_a','mean_d_0_a', 'mean_d_1_a', 'mean_d_2_a', 'mean_d_3_a', 'mean_d_4_a']]
y = df['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=123)
```

```python
X_train.columns
```

```python
print("x_train shape: ", X_train.shape)
```

```python
from sklearn.tree                   import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
DecisionTree_confidence = clf.score(X_test, y_test)
print("DecisionTree: ",DecisionTree_confidence)
```

```python
from sklearn.svm import SVR
svr_rbf = SVR()
svr_rbf.fit(X_train, y_train)
svm_confidence = svr_rbf.score(X_test, y_test)
print("svm confidence: ", svm_confidence)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
clf = GaussianProcessRegressor(random_state=42)
clf.fit(X_train, y_train)
gpr = clf.score(X_test, y_test)
print("GPR: ", gpr)
#y_pred = clf.predict(X)
```

```python
from sklearn.neighbors              import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
knn_confidence = knn.score(X_test, y_test)
print("DecisionTree: ",knn_confidence)
```

```python
y_head_knn = knn.predict(X_test)
y_head_dt  = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm_knn = confusion_matrix(y_test,y_head_knn)
cm_DT = confusion_matrix(y_test,y_head_dt)
```

```python
plt.figure(figsize=(24,12))

plt.suptitle("Confusion Matrixes",fontsize=24)
plt.subplots_adjust(wspace = 0.4, hspace= 0.4)

plt.subplot(2,3,1)
plt.title("Decision tree Confusion Matrix")
sns.heatmap(cm_knn,annot=True,cmap="Blues",fmt="d",cbar=False, annot_kws={"size": 24})
```

```python
plt.subplot(2,3,2)
plt.title("K Nearest Neighbors Confusion Matrix")
sns.heatmap(cm_DT,annot=True,cmap="Blues",fmt="d",cbar=False, annot_kws={"size": 24})
```
